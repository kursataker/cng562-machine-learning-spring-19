```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
def radius_distribution(N=1, size=5000):
  mu = np.zeros(N)
  cov = np.eye(N)
  X = np.random.multivariate_normal(mean=mu, cov=cov, size=size)
  S = np.sum(X*X, axis=1)
  r = np.apply_along_axis(np.sqrt, axis=0, arr=S)
  return r
```

```
rs = [radius_distribution(N) for N in range(1,21)]
```

```
[sns.distplot(r, bins=100) for r in rs]
```
